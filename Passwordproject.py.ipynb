```python
#using while loop
password="Sarthak31"
attempts=3
while attempts>0:
    user_input=input("Enter password:")
    if user_input==password:
        print("Access granted")
        break
    else:
        attempts-=1
        print(f"Wrong password!{attempts} attempts left.")
```
